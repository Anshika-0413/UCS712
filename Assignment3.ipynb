```python
#q6
data = {
    "Employee_ID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Edward"],
    "Department": ["HR", "IT", "IT", "Marketing", "Sales"],
    "Age": [29, 34, 41, 28, 38],
    "Salary": [50000, 70000, 65000, 55000, 60000],
    "Years_of_Experience": [4, 8, 10, 3, 12],
    "Joining_Date": ["2020-03-15", "2017-07-19", "2013-06-01", "2021-02-10", "2010-11-25"],
    "Gender": ["Female", "Male", "Male", "Female", "Male"],
    "Bonus": [5000, 7000, 6000, 4500, 5000],
    "Rating": [4.5, 4.0, 3.8, 4.7, 3.5]
}

df = pd.DataFrame(data)
print(df)
```

```python
#q6(a)
print("Shape of the DataFrame:", df.shape)
```

```python
#q6(b)
df.info()
```

```python
#q6(c)
df.describe()
```

```python
#q6(d)
print(df.head())
print(df.tail(3))
```

```python
#q6(e)
import pandas as pd

# i.
average_salary = df['Salary'].mean()

# ii.
total_bonus = df['Bonus'].sum()

# iii.
youngest_age = df['Age'].min()

# iv.
highest_performance_rating = df['Rating'].max()


print("Average Salary:" ,average_salary)
print("Total Bonus Paid:" ,total_bonus)
print("Youngest Employee's Age:", youngest_age)
print("Highest Performance Rating:",highest_performance_rating)
```

```python
 #q6(f)
 df=df.sort_values(by='Salary', ascending=False)
print(df)
```

```python
#q6(g)
def categorize_performance(rating):
    if rating >= 4.5:
        return 'Excellent'
    elif rating >= 4.0:
        return 'Good'
    else:
        return 'Average'


df['performance_category'] =df['Rating'].apply(categorize_performance)
print(df)
```

```python
#q6(h)
missing_values = df.isnull().sum()


print(missing_values)
```

```python
#q6(i)
df.rename(columns={'Employee_ID': 'ID'}, inplace=True)


print(df)
```

```python
#q6(j)
find_employees = df[(df['Years_of_Experience'] > 5) & (df['Department'] == 'IT')]


print(find_employees)
```

```python
#q6(k)
df['Tax'] = df['Salary'] * 0.10
print(df)
```
